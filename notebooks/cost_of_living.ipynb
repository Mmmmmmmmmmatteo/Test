```python
best_purchasing_power = df.sort_values("Local Purchasing Power Index", ascending = False).iloc[0, :8]
best_purchasing_power
```

```python
import matplotlib.pyplot as plt
top10 = df[["Country", "Local Purchasing Power Index"]].sort_values("Local Purchasing Power Index", ascending=False).head(10)
top10 = top10.set_index("Country")
top10.plot(kind = "barh", figsize=(10,6), color = "teal", legend = False)
plt.xlabel("Top 10 Nazioni per potere d'acquisto")
plt.ylabel("Indice di potere d'acquisto locale")
plt.title("Top 10 Nazioni per Potere d'Acquisto Locale")
plt.gca().invert_yaxis()
```
