### Conclusioni Finali: L'Era della Concentrazione
L'analisi storica dei rendimenti dimostra un chiaro cambio di paradigma nel mercato azionario americano.
Calcolando lo scostamento medio annuo tra l'S&P 500 tradizionale (SPY) e la sua versione Equally Weighted (RSP), emerge un dato inequivocabile:
* Storicamente, i due indici viaggiavano quasi alla pari, indicando una crescita economica diffusa.
* Negli **ultimi 5 anni**, il divario medio annuo  **aumentato di quasi 3 punti percentuali** a favore di SPY.

Questa anomalia statistica conferma che la "bull run" del mercato recente non  dovuta a una salute generale di tutte le 500 aziende americane, ma a un'estrema concentrazione di capitali in pochissimi mega-titoli tecnologici legati all'AI.
